## Task 1 - Collatz Conjecture

>The Collatz conjecture is a famous unsolved problem in maths. The problem is to prove that if you start with any positive integer $x$ and repeatedly apply the function $ f(x) $ below, you always get stuck in the repeating sequence 1,4,2,1,4,2,....

$$ f(x) = {x \div 2} \hspace{1cm} {if\hspace{0.1cm} x\hspace{0.1cm} is\hspace{0.1cm} even} $$


$$              = 3x + 1 \hspace{1cm} otherwise $$

The task is to verify using Python, that the conjecture is true for the first 10,000 integers

```python
# create a function that performs the mathematical operations as outlined above on any positive integer
def f(x):
    # If x is even, divide it by two.
    if x % 2 == 0:
        return x // 2
    else:               # if x is odd, multiply by three and add one.
        return (3 * x) + 1
```

```python
# create a function for the collatz function
def collatz(x):
    print(f'Testing Collatz with initial value {x}')
    if x == 1:         
        print(x, end=". ")
    while x !=1:        # while loop keeps the sequence going while the number is not 1
        x = f(x)
        print(x, end=", ")
    print("\n")
```

```python
for x in range(1, 10001):   # to run the collatz function for all positive integers from one to ten thousand
    if x % 1000 == 0:
        collatz(x)
```

As we can see from the results above, the Collatz conjecture is true for the first 10,000 positive integers.

## Task 4 - Flipping Two Coins

>Suppose you are flipping two coins, each with a probability *p* of giving heads. Plot the entropy of the total number of heads versus *p*.

In information theory, entropy is a measure of uncerainty. It is often referred to as Shannon entropy, after Claude Shannon - an American mathematician anad computer scientist.

Shannon entropy is described by the mathematical function:

<img src="https://miro.medium.com/v2/resize:fit:1400/1*6i3r04xpc-NG_K0LYKU7Sg.png" width="400" height="300"/>

Entropy is a measure of uncerainty. Rolling of two dice are independent events, i.e. knowing the outcome of one roll doesn't tell us anything about the outcome of the other roll.

When rolling two dice there are four possibilities:
- HH (probability = 0.25)
- HT (probability = 0.25)
- TH (probability = 0.25)
- TT (probability = 0.25)

If both dice are fair, the probability of rolling heads on the first dice is 0.5 and the probability of rolling heads on the second dice is 0.5.
Therefore, the proabilities of the total number of heads rolled are:

- 0 heads (TT) = probability of rolling T on first dice and T on the second dice = 0.5 * 0.5 = 0.25
- 1 head (HT or TH) = probability of rolling H on first dice and T on the second dice or T on first dice and H on the second dice = (0.5 * 0.5) + (0.5 * 0.5) = 0.25 + 0.25 = 0.5
- 2 heads (HH) = probability of rolling H on first dice and H on the second dice = 0.5 * 0.5 = 0.25


                                The Entropy or Uncertainty of two events = the sum of the individual Entropies or Uncertainties

$$ H(X,Y) = H(X) + H(Y) $$


Reference = https://towardsdatascience.com/entropy-is-a-measure-of-uncertainty-e2c000301c2c

```python
# define a function in python for Entropy
def H(p):
    return -(1 - p) * np.log2(1.0 - p) - p * np.log2(p)
```

```python
# Entropy of one fair coin
H(0.5)

# Entropy of two fair coins as, from above, the entropy of two events = the sum of the individual entropies
H(0.5) + H(0.5)
```

```python
# Create an empty plot
fig, ax = plt.subplots(figsize=(8,4))

# p is a probability
p = np.linspace(0.00000001, 0.99999999, 1001)

# Plot the Probability vs Entropy of Total No. of Heads When Tossing Two Fair Coins
ax.set_title("Probability vs Entropy of Total No. of Heads When Tossing Two Fair Coins")    # to set the title of the plot
ax.set(xlabel="Probability of Giving Heads", ylabel="Entropy of Total No. of Heads")        # to set the x and y labels for the plot
ax.plot(p, H(p) + H(p));
```

```python
# scatter plot of Culmen Length vs Flipper Length
# to create a scatter plot to visualize the variables using ggplot
# https://r.qcbs.ca/workshop03/book-en/the-basics-of-visualizing-data.html
(ggplot(data = penguins) + aes(x = "bill_length_mm", y = "flipper_length_mm", fill="species") + geom_point() + labs(title = "Relationship between Flipper Length and Body Mass",
subtitle ="Data segmented by species"))
```

```python
# plot Flipper Length vs Body Mass using ggplot
# https://medium.com/@Fortune_/visualizing-the-palmerpenguins-dataset-d3d70bb619b4

(ggplot(data = penguins)+

geom_point(mapping = aes(x="flipper_length_mm", y = "body_mass_g", colour="species"))+

geom_smooth(mapping = aes(x="flipper_length_mm", y = "body_mass_g", colour="species"))+

labs(title = "Relationship between Flipper Length and Body Mass",

subtitle ="Data segmented by species"))
```

### Categorical Variables

####  Species

```python
# to plot a bar chart of the Species variable
x1, counts1= np.unique(species, return_counts=True)

fig, ax = plt.subplots()
bar_colors1 = ['darkorange', 'purple', 'darkcyan']

ax.set_title("Species", fontweight='bold')  # to set the title for the bar chart
ax.set(ylabel="No. of Occurrences")  # to set the y label for the bar chart
ax.bar(x1, counts1, color=bar_colors1)
plt.title('Bar Chart of Species Variable~', fontsize=16);
```
